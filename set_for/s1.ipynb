```python
sdata1 = {'f', 'k', 'r', 'r', 'r', 'k', 'i', 't'}

print(f"sdata1 = {sdata1}, type of sdata1 = {type(sdata1)}, length of sdata1 = {len(sdata1)}")
```

```python
for set in sdata1:
    print(set)
```

```python
sdata1.add('new value')

for set in sdata1:
    print(set)
```

```python
sdata1.remove("new value")
sdata1.discard('something')
```

```python
s1 = {2,6,9,6}
s2 = {2,6,9,8}

print(s1 | s2)

print(f"union = {s1 | s2}")
print(f"intersect = {s1 & s2}")
print(f"difference = {s1 - s2}")
print(f"except = {s2 - s1}")
```
